<font color = "black" > <span style = "font-family : 'Times New Roman' ; font-size: 16pt" >Google ne donne pas beaucoup d'accès aux données sur les requêtes de recherche quotidiennes, mais une autre application de Google connue sous le nom de Google Trends peut être utilisée pour la tâche d'analyse de recherche Google. Google Trends fournit une API qui peut être utilisée pour analyser les recherches quotidiennes sur Google. Cette API est connue sous le nom de pytrends, vous pouvez facilement l'installer dans vos systèmes en utilisant la commande pip ; pip installe pytrends .



```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

<font color = "black" > <span style = "font-family : 'Times New Roman' ; font-size: 16pt" > je vais analyser les tendances de recherche Google sur les requêtes basées sur le « Machine Learning », créons donc un DataFrame des 10 principaux pays qui recherchent le « Machine Learning » sur Google :

```python
trends.build_payload(kw_list=["Machine Learning"])
data = trends.interest_by_region()
data = data.sort_values(by="Machine Learning", ascending=False)
data = data.head(10)
print(data)
```

 <font color = "black" > <span style = "font-family : 'Times New Roman' ; font-size: 16pt" > selon les résultats ci-dessus, les requêtes de recherche basées sur le « Machine learning » sont principalement effectuées à chine. On peut également visualiser ces données à l'aide d'un histogramme :

```python
data.reset_index().plot(x="geoName", 
                        y="Machine Learning", 
                        figsize=(15,12), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

<font color = "black" > <span style = "font-family : 'Times New Roman' ; font-size: 16pt" >Comme nous le savons tous, le Machine Learning a été au centre de l'attention de nombreuses entreprises et d'étudiants au cours des 3-4 dernières années. Examinons donc la tendance des recherches pour voir comment le nombre total de requêtes de recherche basées sur le « Machine Learning » a augmenté. ou diminué sur Google :

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Machine Learning'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Machine Learning', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
